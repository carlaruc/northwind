### Combining tables

We are combining the orders table with the customers table to get more information about each customer

```python
%%sql

SELECT o.order_id, c.company_name, c.contact_name, o.order_date
  FROM orders o
  JOIN customers c
  ON o.customer_id = c.customer_id
  LIMIT 10;
```

Combining orders and employees to see who is responsible for each order

```python
%%sql

SELECT e.first_name ||' '|| e.last_name as employee_name, o.order_id, o.order_date
  FROM employees e
  JOIN orders o
  ON e.employee_id = o.employee_id
  LIMIT 10;
```

Combining order_details, products and orders to get detailed order infomation including the product name and quantity

```python
%%sql

SELECT o.order_id, p.product_name, od.quantity, o.order_date
  FROM orders o
  JOIN order_details od
  ON o.order_id = od.order_id
  JOIN products p
  ON p.product_id = od.product_id
  LIMIT 10;
```

## Ranking Employee Sales Performance

```python
%%sql

WITH employee_sales AS (
  SELECT e.employee_id, first_name, last_name, SUM(unit_price * quantity * (1-Discount)) AS "total_sales"
    FROM employees e
    JOIN orders o
    ON e.employee_id = o.employee_id
    JOIN order_details od
    ON o.order_id = od.order_id
    GROUP BY e.employee_id
)

SELECT employee_id, first_name, last_name, RANK() OVER (ORDER BY "total_sales" DESC) AS "sales_rank"
  FROM employee_sales;
```

### Running Total of Monthly Sales

```python
%%sql
WITH monthly_total AS (
  SELECT DATE_TRUNC('month', order_date)::DATE AS "month", SUM((unit_price * quantity)*(1-discount)) AS "monthly_total_sales"
  FROM orders o
  JOIN order_details od
  ON o.order_id = od.order_id
  GROUP BY DATE_TRUNC('month', order_date))

SELECT "month", SUM(monthly_total_sales) OVER (ORDER BY "month") AS "running_total"
  FROM monthly_total
  ORDER BY "month";
```

### Calculating the % change in sales form one month to the next

```python
%%sql

WITH total_sales_month AS(
    SELECT DATE_TRUNC('month',order_date)::DATE AS "month", SUM(unit_price*quantity*(1-discount)) AS "total_sales"
      FROM order_details od
      JOIN orders o
      ON o.order_id = od.order_id
      GROUP BY DATE_TRUNC('month', order_date)
  ),
  total_sales_previous_month AS(
    SELECT month, total_sales, LAG(total_sales,1) OVER(ORDER BY month) AS previous_month_sales
      FROM total_sales_month
  )

SELECT month, ((total_sales - previous_month_sales)/previous_month_sales) * 100 AS "Growth Rate"
  FROM total_sales_previous_month;
```

### Identifying high value customers

```python
%%sql
WITH customer_value AS(
  SELECT c.customer_id, o.order_id, SUM(unit_price*quantity*(1-discount)) AS order_value
    FROM customers c
    JOIN orders o
    ON c.customer_id = o.customer_id
    JOIN order_details od
    ON o.order_id = od.order_id
    GROUP BY o.order_id, c.customer_id
  )

SELECT customer_id, order_id, order_value,
  CASE
    WHEN order_value > AVG("order_value") OVER() THEN 'Above Average'
    ELSE 'Average/Below Average'
    END AS "Value Category"
  FROM customer_value
  LIMIT 10;
```

### Percentage of sales per category

```python
%%sql
WITH total_sales_product_category AS (
  SELECT c.category_id, c.category_name, SUM(quantity*p.unit_price*(1-discount)) AS "total_sales"
    FROM categories c
    JOIN products p
    ON c.category_id = p.category_id
    JOIN order_details od
    ON od.product_id = p.product_id
    GROUP BY c.category_id
  )

SELECT category_id, category_name, total_sales/SUM(total_sales) OVER()*100 AS "Sales Percentage"
  FROM total_sales_product_category
  ORDER BY "Sales Percentage" DESC;
```

### Top Products Per Category

```python
%%sql

WITH total_sales_product AS(
  SELECT p.product_id, product_name, category_id, SUM(p.unit_price*quantity*(1-discount)) AS "total_sales"
    FROM products p
    JOIN order_details od
    ON p.product_id = od.product_id
    GROUP BY p.product_id, category_id
  )

SELECT product_id, product_name, category_id, total_sales
  FROM (
    SELECT category_id, product_id, product_name, total_sales, 
            ROW_NUMBER() OVER(PARTITION BY category_id ORDER BY total_sales DESC) AS row_number
      FROM total_sales_product) AS tmp
  WHERE row_number <= 3;
```
